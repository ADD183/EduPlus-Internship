```python
#  Global Scope: Configuration and constants
MAX_ATTEMPTS = 3
USERS_DB = {
    "admin": "admin123",
    "research": "bio2026",
    "researcher": "geneAI"
}

# Global variable to track total login sessions
total_logins = 0


def authenticate_user():
    """
    Function to handle user login authentication.
    Demonstrates Local, Enclosing, Nonlocal, and Global variable scopes in a single program.
    """
    global total_logins   # Allows modification of global variable inside the function
    total_logins += 1     # Increment global login count

    print(f"\n Login Attempt #{total_logins}")
    print("Welcome to EduPlusNow Secure Access Portal\n")

    #  Local Scope variable: Tracks if authentication is successful
    is_authenticated = False  

    def login_session():
        """
        Nested function to simulate login attempts.
        Uses 'nonlocal' to modify 'attempts_left' and 'is_authenticated' from enclosing scope.
        """
        #  Enclosing Scope variables
        attempts_left = MAX_ATTEMPTS
        nonlocal is_authenticated

        def attempt():
            """Inner function demonstrating Local Scope"""
            nonlocal attempts_left
            username = input("Enter username: ").strip()
            password = input("Enter password: ").strip()

            if username in USERS_DB and USERS_DB[username] == password:
                print(f"\n Access Granted! Welcome, {username.title()}.\n")
                is_authenticated = True
            else:
                attempts_left -= 1
                print(f" Invalid credentials. Attempts left: {attempts_left}\n")

        # Run login attempts
        while attempts_left > 0 and not is_authenticated:
            attempt()

        # After attempts end
        if not is_authenticated:
            print(" Access Denied. Please try again later.\n")

     # Start user login session
    login_session()

    # After session completion
    if is_authenticated:
        print(" Session Active. Logging activities...")
        user_actions()
    else:
        print(" Login Failed. No session created.")


def user_actions():
    """
    Function representing post-login user operations.
    Demonstrates Local variable usage for task-specific values.
    """
    print("\n Performing operations in secure zone...\n")

    # Local scope variables
    operations = ["View Reports", "Analyze Data", "Logout"]
    for i, op in enumerate(operations, start=1):
        print(f"{i}. {op}")

    choice = int(input("\nSelect an operation (1-3): "))
    if choice == 1:
        print("\n Opening report dashboard...")
    elif choice == 2:
        print("\n Initiating data analysis pipeline...")
    else:
        print("\n Logging out... Goodbye!\n")


# EXECUTION


authenticate_user()

print(f"Total login sessions so far: {total_logins}")
print("\nSystem shutting down securely...")
```

```python
def total_sales(*sales):
    
    """Function to calculate the total sales from multiple branches
    *args allows variables number of inputs
    """

    print("Sales data recieved:",sales)
    total=sum(sales)
    print("Total Sales:", total)
    return total

total_sales(25000,20000,35000)
```

```python
def employee(**details):
    """
    Function to display employee details.
    *kwargs stores all keyword argumentas a dictionary.
    """

    print("\nEmployee Record:")
    for key, value in details.items():
        print((f"{key}:{value}"))
```
