```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

rows = int(input("Enter number of rows: "))

num = 2
for i in range(1, rows + 1):
    count = 0
    while count < i:
        if is_prime(num):
            print(num, end=" ")
            count += 1
        num += 1
    print()
```
