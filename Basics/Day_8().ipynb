```python
#closure

def outer_function(multi):
    def inner_function(value):
        return value * multi
    return inner_function

times3=outer_function(3)
print(times3(5))
```

```python
#Generator

def generator_number():
    for i in range(1,6):
        yield i

gen=generator_number()
for num in gen:
    print(num)
```

```python
#Decorator

def logger(funct):
    def wrapper():
        print(f"Running {funct.__name__}....")
        funct()
        print(f"{funct.__name__} finished!")
    return wrapper

@logger
def greet():
    print("Hello Python Learners!!")
greet()

@logger
def greet2():
    print("Hello Java Learners")
greet2()
```

```python
#Generator Function
def temperature_stream():
    for temp in range(25,45):
        yield temp

#Closure to hold threshold
def threshold_checker(threshold):
    def check(temp):
        return temp>threshold
    return check

#Decorator Function

def logger(func):
    def wrapper(*args,**kwargs):
        print(f"Checking {args[0]} C..")
        result=func(*args,**kwargs)
        if result:
            print(f"Alert: Temperature {args[0]} C exceeds threshold!")
        return result
    return wrapper

@logger
def analyze_temperature(temp):
    checker=threshold_checker(38)
    return checker(temp)

for t in temperature_stream():
    analyze_temperature(t)
```

```python
#Generator Function
def stock_price():
    company={"A":260,"B":254,"C":279,"D":258,"E":298}
    for key, value in company.items():
        yield key,value

#Closure to hold threshold
def threshold_checker(threshold):
    def check(value):
        return value<threshold
    return check

#Decorator Function

def logger(func):
    def wrapper(*args,**kwargs):
        key, value = args
        print(f"Checking {key}:{value}")
        result=func(*args,**kwargs)
        if result:
            print(f"Alert: {key} Company Prices have decreased below threshold !")
            print(f"Prices are:{value}")
        return result
    return wrapper

@logger
def analyze_stock_price(key, value):
    checker=threshold_checker(275)
    return checker(value)

for key, value in stock_price():
    analyze_stock_price(key, value)
```

```python
class student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks

    def show(self):
        print("Name: ",self.name)
        print("Marks: ",self.marks)


s1=student("Arnav",90)
s2=student("Aryan",87)

s1.show()
s2.show()
```

```python
class Bank_Account:

    def __init__(self,account_holder,balance):
        self.account_holder=account_holder
        self.balance=balance

    def deposit(self,amount):
        if amount>self.balance:
            print("Insufficient Balance!!")
        else:
            self.balance=self.balance-amount
            print(f"{amount} withdrawn successfully!")

    def credit(self,amount):
        self.balance+=amount
        print(f"{amount} credited to your account")
        
    def get_balance(self):
        print("Your Balance is:", self.balance)


t1=Bank_Account("Aryan",10000)

print("Account Holder Name: ",t1.account_holder)
print("Initial Balance:",t1.balance)

t1.deposit(2000)
t1.credit(4000)

t1.get_balance()

```
