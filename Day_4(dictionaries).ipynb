```python
dict1={"id":101, "name":"Alice", "Age":30}
print(dict1)

```

```python
print("Student Name:",dict1["name"])
```

```python
#Method-1 for finding key from value
search_value=30
for key, value in dict1.items():
    if value == search_value:
        print("Key for value 30 is:",key)
```

```python
#Method 2
key_with_value=[key for key, value in dict1.items() if value == search_value]
print("Key for value 30 is:",key_with_value)
```

```python
dict2={"a":1, "b":2, "c":3}
dict2["d"]=4
print(dict2)
```

```python
#Method 1 to add new key-value pair
new_value={}
for key, value in dict1.items():
    new_value[key]=value
    if key=="name":
        new_value["city"]="Pune"
print(new_value)
    
```

```python
#Method-2 for finding key from value
dict3={"id":101, "name":"Alice", "Age":30, "Grade":"A"}
new_student={
    **{key: dict3[key] for key in list(dict3)[:2]},
    "city":"Pune",
    **{key: dict3[key] for key in list(dict3)[2:]}
}

print(new_student)
```

```python
#sorting dictionary by keys
unsorted_dict={4:"C", 2:"A", 3:"D", 1:"B"}
sorted_dict=dict(sorted(unsorted_dict.items()))
print("Sorted Dictionary by keys:", sorted_dict)

rev_sorted_dict=dict(sorted(unsorted_dict.items(),reverse=True))
print("Reverse Sorted Dictionary by keys:", rev_sorted_dict)
```

```python
#sorting dictionary based on values
sorted=dict(sorted(unsorted_dict.items(), key=lambda x: x[1]))
print("Sorted Dictionary by values:", sorted)
```
