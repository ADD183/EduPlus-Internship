```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)

plt.plot(x,y)

plt.title("Simple Line Plot of sin(x)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)


plt.plot(x,y1, color="red", linestyle="--", marker="o", label="sin(x)")
plt.plot(x,y2, color="blue", linestyle="--", marker="x", label="cos(x)")

plt.title("Customize Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.grid(True)
plt.legend()

plt.show()
```

```python
categories=["A","B","C","D","E"]
values=[23,25,27,22,29]

plt.bar(categories,values,color='Red')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
colors=np.random.rand(50)
size=1000*np.random.rand(50)

plt.scatter(x,y,c=colors,s=size,alpha=0.5,cmap='viridis')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.colorbar()
plt.show()

```

```python
data=np.random.rand(1000)

plt.hist(data, bins=30,color='orange',edgecolor='black',alpha=0.7)

plt.title("Histogram")
plt.xlabel("Data Range")
plt.ylabel("Frequency")

plt.show()
```

```python
labels=['Apple','Banana','Mango','Kiwi']
sizes=[30,20,25,25]
colors=['Red','Yellow','Orange','Green']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90,shadow=True)

plt.title('Pie Chart Example')
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(x,y1,"r--")
plt.title("siz(x)")

plt.subplot(1,2,2)
plt.plot(x,y2,"b--")
plt.title("cos(x)")

plt.suptitle("Multiple Plots Example")
plt.tight_layout()
plt.show()
```

```python
x=np.linspace(0,5,100)
y=np.exp(x)

plt.plot(x,y,'m--')
plt.title("Exponential Function")
plt.xlabel("X")
plt.ylabel("exp(X)")

plt.savefig("Matplotlib-library-exponential_plot.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
fig,axs= plt.subplots(2,2,figsize=(10,8))

x=np.linspace(0,10,100)
axs[0,0].plot(x,np.sin(x),'r-')
axs[0,0].set_title("Sine Wave")

categories=['A','B','C','D']
values=[10,15,7,12]
axs[0,1].bar(categories,values,color='teal')
axs[0,1].set_title("Bar Chart")

data=np.random.rand(1000)
axs[1,0].hist(data,bins=25,color='purple',edgecolor='black')
axs[1,0].set_title("Histogram")

x=np.random.rand(50)
y=np.random.rand(50)
axs[1,1].scatter(x,y,color='orange')
axs[1,1].set_title("Scatter Plot")

plt.suptitle("Matplotlib Dashboard Examples")
plt.tight_layout()

plt.savefig("Matplotlib-Dashboard.png",dpi=300,bbox_inches='tight')

plt.show()
```
